# Defenders

```python
Defenders2 = Defenders1[Defenders1.MP > 25]
```

```python
Defenders3 = Defenders2.sort_values(by='Tkl+Int', ascending=False)
```

```python
Defenders3.head(n=10)
```

# Midfielders

```python
MF = Player_All.query("position == 'MF'")
```

```python
Midfielders = MF[["Player", "team", "Place", "Status", "position", "Nation", "Age", "MP", "Min", "Gls", "Ast", "G+A", "xAG", "PrgC", "PrgP", "PrgR"]]
```

```python
Midfielders.describe()
```

```python
Top_Midfielders = Midfielders[(Midfielders['Gls'] > 2) & (Midfielders['Ast'] > 2) & (Midfielders['xAG'] > 2) & (Midfielders['PrgC'] > 24) & (Midfielders['PrgP'] > 74) & (Midfielders['PrgR'] > 54)]
```

```python
Top_Midfielders
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x='Ast',y='xAG', data=Top_Midfielders, hue='Player', size='Min')
```

```python
Top_Midfielders2 = Top_Midfielders[["Player", "team", "PrgC", "PrgP", "PrgR"]]
```

```python
TM2 = sns.PairGrid(Top_Midfielders2, hue='Player', palette='hls').map(sns.scatterplot)
TM2.add_legend()
```

# Fowards

```python
FW = Player_All.query("position == 'FW'")
```

```python
Forwards = FW[["Player", "team", "Place", "Status", "position", "Nation", "Age", "MP", "Min", "Gls", "G-PK", "PK", "PKatt", "xG", "npxG"]]
```

```python
Forwards.describe()
```

```python
Forwards1 = Forwards.set_index('team')
```

```python
Forwards2 = Forwards1.sort_values(by='Gls', ascending=False)
```

```python
Forwards2['GlsMin'] = Forwards2['Min'] / Forwards2['Gls']
```

```python
Top_Forwards = Forwards2.head(n=10)
```

```python
Top_Forwards
```

```python
sns.barplot(data=Top_Forwards, y= 'Player', x= 'GlsMin')
```
